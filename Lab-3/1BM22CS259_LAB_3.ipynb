```python

import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('/content/Sales.csv')
x = data['X(weeks)']
y = data['Y(sales)']

# Calculate the mean of x and y
x_mean = x.mean()
y_mean = y.mean()

# Calculate the slope (b1)
numerator = ((x - x_mean) * (y - y_mean)).sum()
denominator = ((x - x_mean)**2).sum()
b1 = numerator / denominator

# Calculate the intercept (b0)
b0 = y_mean - (b1 * x_mean)

# Print the slope and intercept
print("Slope (b1):", b1)
print("Intercept (b0):", b0)

# Predictive model for x = 5
x_new = 5
y_predicted = b0 + (b1 * x_new)

# Print the predicted value
print("Predicted value for x = 5:", y_predicted)
```

```python
import numpy as np
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('/content/Sales.csv')
x = data['X(weeks)']
y = data['Y(sales)']

# Convert x and y to numpy arrays
x = x.values
y = y.values

# Add a column of ones to the input data (for the intercept term)
X = np.ones((len(x), 2))
X[:, 1] = x

# Convert y to a column vector
y = y.reshape(-1, 1)

# Calculate the coefficients using the Normal Equation
# Î¸ = (X^T * X)^(-1) * X^T * y
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# Extract the intercept (b0) and slope (b1)
b0 = theta[0][0]
b1 = theta[1][0]

# Print the slope and intercept
print("Slope (b1):", b1)
print("Intercept (b0):", b0)

# Predictive model for x = 5
x_new = 5
y_predicted = b0 + (b1 * x_new)

# Print the predicted value
print("Predicted value for x = 5:", y_predicted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('/content/Sales.csv')
x = data['X(weeks)']
y = data['Y(sales)']

# Calculate the mean of x and y
x_mean = x.mean()
y_mean = y.mean()

# Calculate the slope (b1)
numerator = ((x - x_mean) * (y - y_mean)).sum()
denominator = ((x - x_mean)**2).sum()
b1 = numerator / denominator

# Calculate the intercept (b0)
b0 = y_mean - (b1 * x_mean)

# Predictive model for x = 5
x_new = 5
y_predicted = b0 + (b1 * x_new)

# Print the slope and intercept
print("Slope (b1):", b1)
print("Intercept (b0):", b0)
print("Predicted value for x = 5:", y_predicted)

# Plot the data points
plt.scatter(x, y, color='blue', label='Actual Data')

# Plot the regression line
y_line = b0 + b1 * x  # Line equation
plt.plot(x, y_line, color='red', label='Regression Line')

# Highlight the predicted point
plt.scatter(x_new, y_predicted, color='green', marker='o', label=f'Prediction (x={x_new})')

# Labels and title
plt.xlabel('X (weeks)')
plt.ylabel('Y (sales)')
plt.title('Linear Regression: Sales vs Weeks')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 5, 9])

# Calculate mean
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate slope (b1) and intercept (b0)
b1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
b0 = y_mean - (b1 * x_mean)

# Generate regression line values
y_line = b0 + b1 * x

# Predictive model for x = 5
x_new = 5
y_predicted = b0 + (b1 * x_new)

# Print results
print("Slope (b1):", b1)
print("Intercept (b0):", b0)
print(f"Predicted value for x = {x_new}: {y_predicted}")

# Plot actual data points
plt.scatter(x, y, color='blue', label='Actual Data', alpha=0.7)

# Plot the regression line
plt.plot(x, y_line, color='red', label='Regression Line')

# Highlight the predicted point
plt.scatter(x_new, y_predicted, color='green', s=100, marker='o', label=f'Prediction (x={x_new})')

# Labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression: Given Data')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
